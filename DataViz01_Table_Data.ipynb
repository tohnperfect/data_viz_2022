# Parallel Coordinates

```python
pd.plotting.parallel_coordinates(df, 'Name')
```

### select columns to show

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength'])
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength']);
```

### change color

https://matplotlib.org/stable/tutorials/colors/colors.html

```python
pd.plotting.parallel_coordinates(df, 'Name', color=['r','y','k'])
```

```python
pd.plotting.parallel_coordinates(df, 'Name', color=[[0.5,0.5,0.5],[0,0,0.8],[0.2,0.4,0.7]])
```

```python
pd.plotting.parallel_coordinates(df, 'Name', color=['#FF00E0', '#00FFE2', '#ECFF00'])
```

### Specify color by colormap
